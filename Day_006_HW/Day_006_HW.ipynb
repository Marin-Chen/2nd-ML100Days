```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = 'D:\\2nd-ML100Days\\2nd-ML100Days\\Day_006_HW\\'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

<h1>請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因</h1>

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('float64')]

numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2 ))])
print("Numbers of remain columns: " ,len(numeric_columns))

# 檢視這些欄位的數值範圍
for col in numeric_columns:
    print(col," 範圍 ： ",app_train[col].min(),"-",app_train[col].max())
    
    app_train[col].hist()
    plt.show()
    app_train[[col]].boxplot(vert=False)
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

```python
amt_income_total = app_train['AMT_INCOME_TOTAL']
print("mean:", amt_income_total.mean())
print("std:", amt_income_total.std())
amt_out = amt_income_total[np.abs(amt_income_total-amt_income_total.mean()) >= (amt_income_total.std()*3)]
amt_in = amt_income_total[np.abs(amt_income_total-amt_income_total.mean()) < (amt_income_total.std()*3)]

print("amt_in:", amt_in.count())
amt_in.hist()
plt.show()

print("amt_out:", amt_out.count())
amt_out.hist()
plt.show()
amt_out.sort_values(ascending=False).head()
```

```python
obs60cnt = app_train['OBS_60_CNT_SOCIAL_CIRCLE']

print("max:", obs60cnt.max())
print("mean:", obs60cnt.mean())
print("std:", obs60cnt.std())
obs60cnt_out = obs60cnt[np.abs(obs60cnt-obs60cnt.mean()) >= (obs60cnt.std()*3)]
obs60cnt_in = obs60cnt[np.abs(obs60cnt-obs60cnt.mean()) < (obs60cnt.std()*3)]

print("obs60cnt_in:", obs60cnt_in.count())
obs60cnt_in.hist()
plt.show()

print("obs60cnt_out:", obs60cnt_out.count())
obs60cnt_out.hist()
plt.show()
obs60cnt_out.sort_values(ascending=False).head()
```

```python
from scipy import stats
outlier_col=[]
for col in numeric_columns:
    x=np.abs(stats.zscore(app_train[col][:]))
    for i in range (0,len(x)):
        if x[i]>3:
            if col not in outlier_col:
                outlier_col.append(col)
                    
print(outlier_col)
```

<h1>Hints: Emprical Cumulative Density Plot, ECDF, ECDF with Python</h1>

```python
def ecdf(data):
    x = np.sort(data)
    y = np.arange(1, len(x)+1) / len(x)
    return (x,y)
```

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
# cdf = app_train['AMT_INCOME_TOTAL'].sort_values().reset_index(drop = True)
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])

# # plt.plot(list(cdf.index), cdf/cdf.max())
plt.plot(x,y)
plt.xlabel('Value')
plt.ylabel('ECDF')
# plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
# plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
# # plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
# plt.plot(np.log(cdf), cdf.index/len(cdf))
x, y = ecdf(app_train['AMT_INCOME_TOTAL'])
plt.xscale('log')
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

# plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
